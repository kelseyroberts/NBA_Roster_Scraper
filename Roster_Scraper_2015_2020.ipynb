```python
import pandas as pd
from requests import get
from bs4 import BeautifulSoup
def get_roster(team, season_end_year):
    r = get(f'https://widgets.sports-reference.com/wg.fcgi?css=1&site=bbr&url=%2Fteams%2F{team}%2F{season_end_year}.html&div=div_roster')
    df = None
    if r.status_code==200:
        soup = BeautifulSoup(r.content, 'html.parser')
        table = soup.find('table')
        df = pd.read_html(str(table))[0]
        df.columns = ['NUMBER', 'PLAYER', 'POS', 'HEIGHT', 'WEIGHT', 'BIRTH_DATE',
                        'NATIONALITY', 'EXPERIENCE', 'COLLEGE']
        df['BIRTH_DATE'] = df['BIRTH_DATE'].apply(lambda x: pd.to_datetime(x))
        df['NATIONALITY'] = df['NATIONALITY'].apply(lambda x: x.upper())
        df['Season']= season_end_year
        df['Gleague_SeasonMatch']= (season_end_year-2) ##since the g league data uses year as beginning and this uses year end, -2 to match
        df['Team']=team
    return df
```

```python
##creating a list of all team abbreviations for loop later
Team_Abr= ['ATL',
'BOS',
'BRK',
'CHI',
'CHO',
'CLE',
'DAL',
'DEN',
'DET',
'GSW',
'PHI',
'HOU',
'IND',
'LAC',
'LAL',
'MIN',
'MEM',
'MIA',
'MIL',
'NOP',
'NYK',
'OKC',
'ORL',
'PHO',
'POR',
'SAC',
'SAS',
'TOR',
'UTA',
'WAS']
```

```python
##creating a list of all the years I want to scrape
##these are using year end
###2014-2015 season is 2015
years=[2015,2016,2017,2018,2019,2020]
```

```python
# using list comprehension  
# to compute all possible permutations 
res = [[i, j] for i in Team_Abr 
              for j in years]
```

```python
##checking to see if it worked
res
```

```python
results=res
```

```python
##using the result combination in the get_roster function to get all rosters for the teams and years I want
###printing the output at every level 
for i in range (0,len(results)):
    print(get_roster(results[i][0],results[i][1]))
```
